## Comparing Police Districts
- Which district has the most crimes? 
- Which has the least?

```python
# get a list of unique districts
district_list = df['District'].unique()
district_list
```

```python
# create an empty dictionary  
districts = {}
# loop through the list of districts
for district in district_list:
    # save a temporary DataFrame of just the rows that match the district
    temp = df.loc[df['District']==district].copy()
    # resample the temporary DataFrame as yearly data of crime counts and keep only the .size() 
    temp_res = temp.resample("Y").size() 
    # save the temporary DataFrame in the dictionary using the district as the key
    districts[district] = temp_res.copy() 
districts.keys()
```

```python
# convert the dictionary to a dataframe
district_df = pd.DataFrame(districts)
pd.set_option("display.max_columns", 100)
district_df
```
