### Convert the date column to datetime dtype

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date']
```

### Set the date column as the index

```python
df = df.set_index('Date')
df.head()
```

### Keep 2 forms of the data

#### The original individual crime data with a datetime index (each row is 1 crime)

```python
df_crime = df.copy()
df_crime.head()
```

#### A resampled/converted crime counts version (each row is 1 day)

```python
df_day = df.copy()
df_day = df_day.reset_index()
df_day['year'] = df_day['Date'].dt.year
df_day['month'] = df_day['Date'].dt.month_name()
df_day['day of month'] = df_day['Date'].dt.day
df_day['day of week'] = df_day['Date'].dt.day_name()
df_day['hour'] = df_day['Date'].dt.hour
df_day['Count'] = 1
df_day
```

```python
df_day['Date'] = pd.to_datetime(df.index.strftime('%Y-%m-%d'))
df_day
```

## Comparing Police Districts
- Which district has the most crimes? 
- Which has the least?

```python
df_crime['District'].value_counts()
```

- District 8.0 has the most crimes (521830).
- District 21.0 has the least crimes (4).

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax = sns.countplot(data = df_crime, y = 'District', palette = 'viridis')
ax.set(title = 'Crime per Police District')
ax.axvline(521830, color = 'k', lw = 2, ls = ':', label = 'District 8.0 - 521,830 crimes')
ax.axvline(4, color = 'r', lw = 5, ls = ':', label = 'District 21.0 - 4 crimes')
ax.legend();
```
