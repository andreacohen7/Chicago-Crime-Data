## Crimes Across the Years


### Is the total number of crimes increasing or decreasing across the years?

```python
# resample the data
df_year1 = df_day.resample('A')['Count'].sum().reset_index()
df_year1
```

```python
# set the date column as the index
df_year1 = df_year1.set_index('Date')
df_year1
```

```python
# calculate the difference from year to year
ts_diff = df_year1.diff()
ts_diff
```

```python
# remove year 2001 (NaN) and 2023 (not a full year's data yet)
ts_diff = ts_diff['2002-12-31':'2022-12-31']
```

```python
# plot the change in number of crimes 
ax = ts_diff.plot(figsize = (10,3), ylabel = 'Change in Number of Crimes', title = 'Change in Number of Crimes from 2002-2022')
ax.axhline(0, color = 'k');
```

- The total number of crimes decreased from 2002 to 2015. 
- They increased from 2015 to 2016, then decreased again until 2021.
- From 2021 to 2022 the total number of crimes increased again.

### Are there any individual crimes that are doing the opposite (e.g decreasing when overall crime is increasing or vice-versa)?

```python
# resample the data
df_year2 = df_day.groupby('Primary Type').resample('A')['Count'].sum().reset_index()
df_year2
```

```python
# pivot the result to have the datetime as the index and the crime as the columns
years_ts = df_year2.pivot(index = 'Date', columns = 'Primary Type')
years_ts
```

```python
# drop the first level of columns
years_ts.columns = years_ts.columns.droplevel(0)
years_ts
```

```python
# subtract to get delta crimes
delta_crime_years = years_ts.loc['2022-12-31'] - years_ts.loc['2001-12-31']
delta_crime_years
```

```python
# sort the values of delta crimes
delta_crime_years.sort_values()
```

```python
# drop the missing values in delta crimes
delta_crime_years = delta_crime_years.dropna()
```

```python
# check that the missing values were dropped
delta_crime_years.sort_values()
```

```python
# plot the changes
ax = delta_crime_years.sort_values().plot(kind = 'bar', figsize = (10,3), ylabel = 'Change in # Crimes', title = 'Change in Crimes from 2001-2022')
ax.axhline(0 , color = 'k');
```

- Overall crime decreased, but other narcotic violation, obscenity, homicide, concealed carry license violation, stalking, deceptive practice, criminal sexual assault, and weapons violation all increased.

## Comparing AM vs. PM Rush Hour

### Are crimes more common during AM rush hour or PM rush hour?
- AM rush hour = 7am-10am
- PM rush hour = 4pm-7pm

```python
# number of crimes per hour
hour_counts = df_day['hour'].value_counts().sort_index()
hour_counts
```

```python
# number of crimes during morning rush hour
am_hour_counts = hour_counts[7:10]
display(am_hour_counts)
am_hour_counts.sum()
```

```python
# number of crimes during evening rush hour
pm_hour_counts = hour_counts[16:19]
display(pm_hour_counts)
pm_hour_counts.sum()
```

```python
# plot the number of crimes
fig, axes = plt.subplots(ncols = 2, figsize = (12,4))
am_hour_counts.plot(style = 'o-', ylabel = 'Number of Crimes', xlabel = 'AM Rush Hour', xticks = am_hour_counts.index, ax = axes[0])
pm_hour_counts.plot(style = 'o-', ylabel = 'Number of Crimes', xlabel = 'PM Rush Hour', xticks = pm_hour_counts.index, ax = axes[1])
axes[0].set_ylim(bottom=15000, top=450000)
axes[1].set_ylim(bottom=15000, top=450000)
fig.suptitle('Crime During Rush Hour')
fig.tight_layout();
```

- Crimes are more common during PM rush hour than AM rush hour.

### What are the top 5 most common crimes during AM rush hour? 

```python
crimes_by_hour = df_day.groupby('hour')['Primary Type'].value_counts(normalize = False)
crimes_by_hour
```

```python
amrushhour_crimes = crimes_by_hour.loc[7] + crimes_by_hour.loc[8] + crimes_by_hour.loc[9]
amrushhour_crimes
```

```python
ax = amrushhour_crimes.head().sort_values().plot(kind = 'barh', figsize = (10, 3))
ax.set(title = 'Most Common Crimes for AM Rush Hour', xlabel = 'Number of Crimes');
```

- The 5 most common crimes during AM rush hour are battery, burglary, assault, arson, and concealed carry license violation.

### What are the top 5 most common crimes during PM rush hour?

```python
pmrushhour_crimes = crimes_by_hour.loc[16] + crimes_by_hour.loc[17] + crimes_by_hour.loc[18]
pmrushhour_crimes
```

```python
pmrushhour_crimes = pmrushhour_crimes.dropna()
pmrushhour_crimes.head()
```

```python
ax = pmrushhour_crimes.head().sort_values().plot(kind = 'barh', figsize = (10, 3))
ax.set(title = 'Most Common Crimes for PM Rush Hour', xlabel = 'Number of Crimes');
```

- The 5 most common crimes during PM rush hour are battery, assault, burglary, arson, and concealed carry license violation.

### Are Motor Vehicle Thefts more common during AM rush hour or PM Rush Hour?

```python
amrushhour_crimes['MOTOR VEHICLE THEFT']
```

```python
pmrushhour_crimes['MOTOR VEHICLE THEFT']
```

- Motor vehicle theft is more common during PM rush hour than during AM rush hour.

```python
rh_df = df_day.copy()
# filters for group conditions
amrushhour_start = rh_df['hour'] >= 7
amrushhour_end = rh_df['hour'] < 10
pmrushhour_start = rh_df['hour'] >= 16
pmrushhour_end = rh_df['hour'] < 19
# use filters to add group column
rh_df.loc[amrushhour_start & amrushhour_end, 'Group'] = 'AM Rush Hour'
rh_df.loc[pmrushhour_start & pmrushhour_end, 'Group'] = 'PM Rush Hour'
# filter for motor vehicle theft
mvt = rh_df['Primary Type'] == 'MOTOR VEHICLE THEFT'
mvt_df = rh_df.loc[mvt]
mvt_df['Group'].value_counts()
```

```python
fig, ax = plt.subplots (figsize = (10, 5))
ax = sns.countplot(data = mvt_df, x = 'Group', order = ['AM Rush Hour', 'PM Rush Hour'])
ax.set(title = 'Motor Vehicle Theft During AM and PM Rush Hour', ylabel = 'Motor Vehicle Theft');
```
